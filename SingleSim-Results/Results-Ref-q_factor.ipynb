# Reference Population: low *vs* high q_factor

The parameters are summarized in the table below (**N = Neutral, S = Selection**):

Parameter  |Low q (N/S)|High q(N/S)|
-----------|---------|------------|
Final size | $10^6$  |   $10^6$   |
q (start)  |$10^{-3}$| $10^{-3}$  |
**q-factor**   | $1$   |   $5$    |  
P(death)   | $0.3$   |   $0.3$    |
Selection  | 0 / 0.3 |  0 / 0.3   |

# Simulations

## Low q_factor (Neutral)

```python
# set seed to get same result for each test simulation
np.random.seed(123)

# Test cell
l = 10**6
q = 10**(-3)
q_fac = 1
alpha = 0.3
sel = 0
low_qfac_N = Population(l, q, mutfactor=q_fac, death_frac=alpha, selection=sel)
low_qfac_N.simulate()
print("Total population size:", low_qfac_N.size)
print("Generations:", low_qfac_N.gen)
print()
print("Start clone size:", low_qfac_N.start_clone.size)
print("Number of subclones:", len(low_qfac_N.clones[1:]))
```

## Low q_factor (Selection)

```python
# set seed to get same result for each test simulation
np.random.seed(123)

# Test cell
l = 10**6
q = 10**(-3)
q_fac = 1
alpha = 0.3
sel = 0.3
low_qfac_S = Population(l, q, mutfactor=q_fac, death_frac=alpha, selection=sel)
low_qfac_S.simulate()
print("Total population size:", low_qfac_S.size)
print("Generations:", low_qfac_S.gen)
print()
print("Start clone size:", low_qfac_S.start_clone.size)
print("Number of subclones:", len(low_qfac_S.clones[1:]))
```

## High q_factor (Neutral)

```python
# set seed to get same result for each test simulation
np.random.seed(123)

# Test cell
l = 10**6
q = 10**(-3)
q_fac = 5
alpha = 0.3
sel = 0
high_qfac_N = Population(l, q, mutfactor=q_fac, death_frac=alpha, selection=sel)
high_qfac_N.simulate()
print("Total population size:", high_qfac_N.size)
print("Generations:", high_qfac_N.gen)
print()
print("Start clone size:", high_qfac_N.start_clone.size)
print("Number of subclones:", len(high_qfac_N.clones[1:]))
```

## High q (Selection)

```python
# set seed to get same result for each test simulation
np.random.seed(123)

# Test cell
l = 10**6
q = 10**(-3)
q_fac = 5
alpha = 0.3
sel = 0.3
high_qfac_S = Population(l, q, mutfactor=q_fac, death_frac=alpha, selection=sel)
high_qfac_S.simulate()
print("Total population size:", high_qfac_S.size)
print("Generations:", high_qfac_S.gen)
print()
print("Start clone size:", high_qfac_S.start_clone.size)
print("Number of subclones:", len(high_qfac_S.clones[1:]))
```

```python
# Check weights of clones for population with selection
low_qfac_weights = []
for clone in low_qfac_S.clones:
    low_qfac_weights.append(clone.weight)

high_qfac_weights = []
for clone in high_qfac_S.clones:
    high_qfac_weights.append(clone.weight)

plt.figure();

plt.subplot(121);
plt.hist(low_qfac_weights, bins=100, color='r', alpha=0.6);
plt.title("Distribution of weights of low_qfac_S");
plt.xlabel("Weight");
plt.ylabel("Frequency (clones)");

plt.subplot(122);
plt.hist(high_qfac_weights, bins=100, color='g', alpha=0.6);
plt.title("Distribution of weights of high_qfac_S");
plt.xlabel("Weight");
plt.ylabel("Frequency (clones)");

plt.show()
```

## Distribution of number of mutations for all Populations

```python
low_qfac_N_num_mut_df = num_mutations_distribution(low_qfac_N, det_lim=detection_limit)
low_qfac_N_num_mut_df.columns = ["qfac = 1, Neutral evolution"]
#low_qfac_N_num_mut_df.plot(kind='bar', logy=True, legend=True, rot=0, color='b', alpha=0.6);

low_qfac_S_num_mut_df = num_mutations_distribution(low_qfac_S, det_lim=detection_limit)
low_qfac_S_num_mut_df.columns = ["qfac = 1, Selection = %s" %(sel)]
#low_qfac_Snum_mut_df.plot(kind='bar', logy=True, legend=True, rot=0, color='r', alpha=0.6);

high_qfac_N_num_mut_df = num_mutations_distribution(high_qfac_N, det_lim=detection_limit)
high_qfac_N_num_mut_df.columns = ["qfac = 5, Neutral evolution"]
#low_qfac_N_num_mut_df.plot(kind='bar', logy=True, legend=True, rot=0, color='b', alpha=0.6);

high_qfac_S_num_mut_df = num_mutations_distribution(high_qfac_S, det_lim=detection_limit)
high_qfac_S_num_mut_df.columns = ["qfac = 5, Selection = %s" %(sel)]
#low_qfac_N_num_mut_df.plot(kind='bar', logy=True, legend=True, rot=0, color='b', alpha=0.6);

combo_low = low_qfac_N_num_mut_df.join(low_qfac_S_num_mut_df, how='right')
combo_high = high_qfac_N_num_mut_df.join(high_qfac_S_num_mut_df, how='right')
combo = combo_low.join(combo_high, how='right')

combo.plot(kind='bar', logy=True, legend=True, rot=0, alpha=0.6, );
plt.ylabel("Cell count");
plt.title("Number of mutations distribution");
plt.savefig("./Figures/Ref-q_factor/q_factor-Num_mut.png")
```
