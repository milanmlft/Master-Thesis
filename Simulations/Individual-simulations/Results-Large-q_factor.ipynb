# Simulations

## High q_factor (Neutral)

```python
%%time
# set seed to get same result for each test simulation
np.random.seed(123)

# Test cell
l = 10**8
q = 10**(-3)
q_fac = 3
alpha = 0.3
sel = 0
high_qfac_N = Population(l, q, mutfactor=q_fac, death_frac=alpha, selection=sel)
high_qfac_N.simulate()
print("Total population size:", high_qfac_N.size)
print("Generations:", high_qfac_N.gen)
print()
print("Start clone size:", high_qfac_N.start_clone.size)
print("Number of subclones:", len(high_qfac_N.clones[1:]))
```

## High q (Selection)

```python
%%time
# set seed to get same result for each test simulation
np.random.seed(123)

# Test cell
l = 10**8
q = 10**(-3)
q_fac = 3
alpha = 0.3
sel = 0.3
high_qfac_S = Population(l, q, mutfactor=q_fac, death_frac=alpha, selection=sel)
high_qfac_S.simulate()
print("Total population size:", high_qfac_S.size)
print("Generations:", high_qfac_S.gen)
print()
print("Start clone size:", high_qfac_S.start_clone.size)
print("Number of subclones:", len(high_qfac_S.clones[1:]))
```

```python
# Check weights of clones for population with selection
high_qfac_weights = []
for clone in high_qfac_S.clones:
    high_qfac_weights.append(clone.weight)

plt.hist(high_qfac_weights, bins=100, color='g', alpha=0.6);
plt.title("Distribution of weights of high_qfac_S");
plt.xlabel("Weight");
plt.ylabel("Frequency (clones)");

plt.show()
```

## Distribution of number of mutations for high_qfac Populations

```python
high_qfac_N_num_mut_df = num_mutations_distribution(high_qfac_N_data, det_lim=detection_limit)
high_qfac_N_avg_num_mut = avg_num_mutations(high_qfac_N_num_mut_df)
high_qfac_N_num_mut_df.columns = ["Neutral, mean = %s" %(round(high_qfac_N_avg_num_mut, 3))]
#high_qfac_N_num_mut_df.plot(kind='bar', logy=True, legend=True, rot=0, color='b', alpha=0.6);

high_qfac_S_num_mut_df = num_mutations_distribution(high_qfac_S_data, det_lim=detection_limit)
high_qfac_S_avg_num_mut = avg_num_mutations(high_qfac_S_num_mut_df)
high_qfac_S_num_mut_df.columns = ["Selection = %s, mean = %s" %(sel, round(high_qfac_S_avg_num_mut, 3))]
#high_qfac_S_num_mut_df.plot(kind='bar', logy=True, legend=True, rot=0, color='r', alpha=0.6);

combo = high_qfac_N_num_mut_df.join(high_qfac_S_num_mut_df, how='right')
print(combo)
combo.plot(kind='bar', logy=True, legend=True, rot=0, color=['b', 'r'], alpha=0.6);

plt.ylabel("Cell count");
plt.xlabel("Number of mutations");
plt.legend(frameon=True, fancybox=True, shadow=True);
plt.savefig("./Figures/Large-q_factor/q_factor-Num_mut.png")
plt.show()
```
